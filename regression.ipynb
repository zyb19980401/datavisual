```python
Geodata_clean_bypo = gpd.sjoin(Geodata_clean, zips, how='left')
Geodata_clean_bypo.reset_index()
Geodata_clean_bypo.head()
```

```python
Geodata_clean_bypo = Geodata_clean_bypo[['CMPLNT_NUM', 'OFNS_DESC', 'BORO_NM', 'PO_NAME', 'COUNTY',"in_area", 'POPULATION']]
Geodata_clean_bypo.head()
```

```python
def convert(nparray, oldboro_2_new):
    result = []
    for boro in nparray:
        
        result.append(oldboro_2_new[boro])
    return result
```

```python
Geodata_clean_bypo["COUNTY"] = convert(Geodata_clean_bypo["COUNTY"], oldboro_2_new)
Geodata_clean_bypo = Geodata_clean_bypo[Geodata_clean_bypo["COUNTY"] == Geodata_clean_bypo["BORO_NM"]]
Geodata_clean_bypo.head()
```

```python
Geodata_bytypenpo = Geodata_clean_bypo.groupby(['OFNS_DESC', 'PO_NAME', 'in_area'], as_index=False).agg({'CMPLNT_NUM':'count', 'POPULATION':'sum'})
Geodata_bytypenpo['norm_CMPLNT_NUM'] = 100000 * Geodata_bytypenpo['CMPLNT_NUM'] / Geodata_bytypenpo['POPULATION'] 
Geodata_bytypenpo.head()
```

```python
zip_popbybor = zips.groupby(by='COUNTY',as_index=False).agg({'POPULATION':'sum'})
zip_popbybor["COUNTY"] = convert(zip_popbybor["COUNTY"], oldboro_2_new)
zip_popbybor.head()
```

```python
Geodata_bytypenboro = Geodata_clean_bypo.groupby(['OFNS_DESC', 'COUNTY', 'in_area'], as_index=False).agg({'CMPLNT_NUM':'count'})
norm_CMPLNT_NUM = []
for i in range(len(Geodata_bytypenboro)):
    bor = Geodata_bytypenboro['COUNTY'][i]
    norm = 100000 * Geodata_bytypenboro['CMPLNT_NUM'][i] / zip_popbybor[zip_popbybor['COUNTY']==bor].reset_index()['POPULATION'][0]
    norm_CMPLNT_NUM.append(norm)
Geodata_bytypenboro['norm_CMPLNT_NUM'] = norm_CMPLNT_NUM
Geodata_bytypenboro.head(20)
```

```python
ttest = []
ktest = []
for boro_1 in boroughs:
    ttest_col = []
    ktest_col = []
    for boro_2 in boroughs:
        ttest_value = stats.ttest_ind(Geodata[Geodata['BORO_NM']==boro_1]['norm_CMPLNT_NUM'], Geodata[Geodata['BORO_NM']==boro_2]['norm_CMPLNT_NUM'])
        ktest_value = stats.ks_2samp(Geodata[Geodata['BORO_NM']==boro_1]['norm_CMPLNT_NUM'], Geodata[Geodata['BORO_NM']==boro_2]['norm_CMPLNT_NUM'])
        ttest_col.append(ttest_value[1])
        ktest_col.append(ktest_value[1])
    ttest.append(ttest_col)
    ktest.append(ktest_col)
```
